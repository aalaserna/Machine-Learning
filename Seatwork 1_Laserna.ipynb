```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
data = pd.read_csv("data/movie_metadata.csv")
data.head(4)
```

<h2>Top 5 Directors with most movies directed</h2>

```python
cols = ['movie_title','director_name']
data[cols]
```

```python
movies_per_director = data.groupby("director_name").size()
movies_per_director
```

```python
top5_by_director = movies_per_director.sort_values(ascending = False)
top5_by_director[:5]
```

<h2>Top 5 Directors ranked based on the total gross incomes from all the movies they directed</h2>

```python
cols = ['gross', 'movie_title', 'director_name']
data[cols]
```

```python
gross_per_director = data.groupby("director_name")['gross'].sum()
gross_per_director
```

```python
top5_by_gross = gross_per_director.sort_values(ascending = False)
top5_by_gross[:5]
```

<h2>How many movies has a movie duration between 150 and 180</h2>

```python
data.columns
```

```python
cols = ['duration', 'movie_title']
data[cols]
```

```python
cond = data['duration'].between(150,180)
movies = data[cond]
total_movies = movies.shape[0]
total_movies
```

<h2>Which country has the most number of movies with duration between 150 and 180</h2>

```python
cols = ['country', 'movie_title']
movies[cols]
```

```python
country = movies[cols].groupby('country').size().sort_values(ascending=False)
country[:1]
```
